##**Passenger Satisfaction**##


```python
plt.figure(figsize=(8,5))
sns.countplot(data=Airplane_cus, x="Satisfaction", palette="coolwarm")
plt.title("Passenger Satisfaction Distribution")
plt.xlabel("Satisfaction Level")
plt.ylabel("Count")
plt.show()
```

The chart reveals that a higher number of airline passengers fall into the "Neutral or Dissatisfied" category compared to those who are "Satisfied." This suggests potential issues with customer experience, service quality, or other factors impacting passenger satisfaction. Airlines can use this insight to identify areas for improvement and enhance overall customer experience.

## SATISFACTION BY AGE DISTRIBUTION ##


```python
plt.figure(figsize=(8,5))
sns.boxplot(data=Airplane_cus, x="Satisfaction", y="Age", palette="coolwarm")
plt.title("Age Distribution by Satisfaction Level")
plt.xlabel("Satisfaction Level")
plt.ylabel("Age")
plt.show()
```

Neutral or Dissatisfied spans ages from 10 to 80, showing a wider spread. On the other hand, the Satisfied group is tighter, from 20 to 70, with a few outliers over 80

## SATISFACTION BY FLIGHT CLASS ##


```python
plt.figure(figsize=(8,5))
sns.countplot(data=Airplane_cus, x="Class", hue="Satisfaction", palette="coolwarm")
plt.title("Satisfaction by Class")
plt.xlabel("Travel Class")
plt.ylabel("Count")
plt.legend(title="Satisfaction")
plt.show()
```

Economy Class has a higher number of dissatisfied passengers. Economy Plus has the lowest number of passengers overall, with more dissatisfaction than satisfaction. This suggests that better service and amenities in Business Class contribute to higher satisfaction, while Economy Class may need improvements to enhance the passenger experience.

## CORRELATION HEATMAP ##

```python
plt.figure(figsize=(12,6))
numeric_features = Airplane_cus.select_dtypes(include=np.number)
sns.heatmap(numeric_features.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

Strong correlation between Departure Delay and Arrival Delay (0.97), Longer departure delays almost always result in arrival delays.

Seat Comfort, Leg Room Service, and Cleanliness are positively correlated, These factors likely contribute to overall passenger comfort and satisfaction.

In-flight Service, Food & Drink, and In-flight Entertainment are moderately correlated ,Suggesting that better services lead to a better experience.

Baggage Handling has a moderate correlation with Check-in Service and Online Boarding , Efficient ground services improve overall passenger experience
